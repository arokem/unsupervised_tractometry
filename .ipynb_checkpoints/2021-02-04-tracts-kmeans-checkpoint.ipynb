## PCA:we form a 20*7700 matrix which have nodes for each tract.

```python

tract = arr.reshape((7700,20))
```

```python
tracts = tract.transpose()
```

```python
tracts.shape
```

```python
tracts = imputer.fit_transform(tracts)
np.sum(np.isnan(tracts))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
tracts_nor = scaler.fit_transform(tracts)
```

```python
pca = PCA(n_components=3)
tracts_pca = pca.fit_transform(tracts_nor)
plt.scatter(tracts_pca[:, 0], tracts_pca[:, 1])
plt.title('PCA(2 components) for FA')
plt.xlabel('component 1')
plt.ylabel('component 2')
```

```python

plt.scatter(tracts_pca[:, 0], tracts_pca[:, 2])
plt.title('PCA(2 components) for FA')
plt.xlabel('component 0')
plt.ylabel('component 2')
```

```python
plt.scatter(tracts_pca[:, 1], tracts_pca[:, 2])
plt.title('PCA(2 components) for FA')
plt.xlabel('component 1')
plt.ylabel('component 2')
```
